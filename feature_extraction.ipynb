```python
import cv2
import matplotlib.pyplot as plt
```

```python
#convert image to pixel, uses 'imread'library
before_img=cv2.imread('before.png',cv2.IMREAD_GRAYSCALE)
before_img
```

```python
before_img.shape
```

```python
#convert pixel to image, here we use 'imshow' library
plt.imshow(before_img,cmap='gray')
```

```python
img_resize=cv2.resize(before_img,(28,28),interpolation=cv2.INTER_LINEAR)
img_resize
```

```python
#standard feature 28,28(28*28)
img_resize.shape
```

```python
plt.imshow(img_resize,cmap='gray')
```

```python
modified_image = cv2.imread('modified.png', cv2.IMREAD_GRAYSCALE)
modified_image
```

```python
modified_image.shape
```

```python
plt.imshow(modified_image,cmap='gray')
```

```python
#'bitwise_not convert color 
img_resizedM = cv2.bitwise_not(modified_image)
img_resizedM
```

```python
img_resizedM.shape
```

```python

plt.imshow(img_resizedM, cmap='gray')
```

```python
again_img_resizedM=cv2.resize(img_resizedM,(28,28),interpolation=cv2.INTER_LINEAR)
again_img_resizedM
```

```python
plt.imshow(again_img_resizedM,cmap='gray')
```
